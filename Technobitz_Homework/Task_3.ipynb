```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
df["size"].unique()
```

```python
male = df[df["sex"] == "Male"]
female = df[df["sex"]== "Female"]
male.plot("total_bill", "tip", kind = "scatter", c = "r")
female.plot("total_bill", "tip", kind = "scatter", c ="b")
```

```python
male.plot("total_bill", kind = "kde")
```

```python
sns.countplot( x = "time", data = df, hue = "size")
```

```python
sns.catplot( x = "sex", y = "tip", hue = "time", data = df )
```

```python
sns.violinplot(x = "time", y = "tip", data = df, hue = "sex")
```

```python
sns.pairplot(data = df, hue = "time")
```

```python
sns.jointplot("total_bill", "tip", data=df, kind="hex", height = 10)
```

```python
sns.lmplot("total_bill", "tip", col="sex", hue="sex", data=df)
```
